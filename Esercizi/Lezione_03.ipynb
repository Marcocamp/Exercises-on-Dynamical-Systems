# Modello di evoluzione di una popolazione isolata

```python
# import numpy as np                    # non serve se non usi array
import control as ct
import matplotlib.pyplot as plot

b, d= 2 , 1

A = b-d
B = 0
C = 1
D = 0

X0=1



sys = ct.ss(A, B, C, D, outputs='popolazione', name='modello popolazione')
results = ct.initial_response(sys, X0 = X0)
grafico = results.plot()
```

```python
import numpy as np

t = np.arange(0,10,0.1)
u = np.zeros(t.shape)

results2 = ct.forced_response(sys, t, u, X0=X0)
plot = results2.plot()
```

```python
A = [[0, 1], [1, 1]]
B = [[0], [-1]]
C = [[1, 1]]
D = [[0]]

sys2 = ct.ss(A, B, C, D, outputs='Galline nel pollaio', name='modello pollaio', inputs='prelevate per la vendita', states=['galline', 'pulcini'],)
t = np.arange(0,10,1)
u = np.ones(t.shape)
results3 = ct.forced_response(sys2, t, u, X0=[100, 50])

grafico2 = results3.plot()
```

```python
t = np.arange(0,1000,0.1)
results4 = ct.frequency_response(sys2)
grafico3 = results4.plot()
```
