```python
import numpy as np
from numpy.polynomial import Polynomial
import sympy as sp
import matplotlib.pyplot as plt
import control as ctrl

# Matrice di stato F
F = sp.Matrix([[2, 1],
              [-1, 0]])

# Nessun ingresso: sistema autonomo
A = F                   # Matrice di stato
B = np.zeros((2, 1))    # Matrice di ingresso
C = np.eye(2)           # Matrice di uscita
D = np.zeros((2, 1))    # Matrice ingresso-uscita

# Sistema nello spazio di stato
sys = ctrl.ss(A, B, C, D)

# Autovalori di F
eigenvalues = sp.Matrix(A).eigenvals()
#eigenvalues = np.linalg.eigvals(F)
print("Autovalori di F:", eigenvalues)

# Polinomio caratteristico
print(A.charpoly())
char_poly = A.charpoly().as_expr()
# char_poly = Polynomial.fromroots(eigenvalues)
print("Polinomio caratteristico:", char_poly)

P, J = F.jordan_form()

print("\nForma di Jordan J:")
sp.pprint(J)

print("\nVerifica: P * J * P^(-1) = F")
if P * J * P**-1 == F:
    print("Verifica riuscita!")
else:
    print("Verifica fallita.")
```
